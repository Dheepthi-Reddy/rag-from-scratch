Step 1: Loading the embedding model

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
print('Model loaded')
```

Step 2: Creatnig Documents

```python
documents=[
    "Built a production RAG pipeline using LangChain, FAISS, and Claude via AWS Bedrock to enable semantic search across 100K+ compliance documents.",
    "Extended RAG platform with multi-step query rewriting, hybrid retrieval, and agentic workflows using LangGraph.",
    "Evaluated answer quality using LLM-as-judge evaluation frameworks on AWS Bedrock.",
    "Built streaming fraud scoring pipelines using Apache Kafka and Spark Structured Streaming serving predictions via SageMaker endpoints.",
    "Optimized GenAI inference through semantic caching, model benchmarking, and prompt engineering on AWS Bedrock."
]

print(f"We have {len(documents)} documents")
print("\nFirst document:",documents[0])
print("Resume bullets encoded")
```

Step 3: Converting documents to vectors

```python
import numpy as np

# Converting documents to vectors
document_vectors = model.encode(documents)

print(f"Shape: {document_vectors.shape}")
print("\nFirst document vector (first 10 numbers):", document_vectors[0][:10])
```

shape: (5, 384)
5 = no of documents
384 = number of dimensions in each vector, (each sentence is represented as 384 numbers)

Step 4: Convert a question to a vector and find the most similar document

```python
# question asked by an analyst
question = "Experience with LLMs and retrieval augmented generation systems"

# converting question to vector
question_vector = model.encode([question])

# Calculating similarity between question and every document
similarities = np.dot(document_vectors, question_vector.T).flatten()

print("Similarity scores:")
for i, score in enumerate(similarities):
  print(f" [{score:.3f}] {documents[i]}")

best_idx = np.argmax(similarities)
print(f"\nMost relevant document: ", documents[best_idx])
```

```python
question_two = "Real time data streaming and ML model deployment at scale"

question_two_vector = model.encode([question_two])
similarities = np.dot(document_vectors, question_two_vector.T).flatten()

print("Relevance scores:")
for i, score in enumerate(similarities):
  print(f" [{score:.3f}] {documents[i]}")

best_idx = np.argmax(similarities)
print(f"Most relevant document: ", documents[best_idx])
```

# RAG

This notebook demonstrates the core of Retrieval Augumented Generation(RAG):

**R - Retrieval:** We convert the question to a vector and search for the most simmilar document using dot product similarity. We retrieve the most relevant context.

**A - Augemented:** Instead of just sending the question to an LLM, we AUGMENT the prompt with the retrieved document as context. This grounds the LLM's answer in real data.

**G - Generation:** The LLM generates an answer based on the retrieved context, not from its training data. This prevents hallucination.

Without retrieval , the LLM gusses. With retrieval, it answers from facts.
